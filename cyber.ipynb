```python
# Extract day, month, and hour from the Timestamp
new_Train['Day'] = new_Train['Timestamp'].dt.day
new_Train['Month'] = new_Train['Timestamp'].dt.month
new_Train['Hour'] = new_Train['Timestamp'].dt.hour
new_Train['Year'] = new_Train['Timestamp'].dt.year


# Drop the original Timestamp column
new_Train.drop('Timestamp', axis=1, inplace=True)

print(new_Train.head())
```

```python
numerical_df = new_Train.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numerical_df.corr().abs()

# Plot the heatmap
plt.figure(figsize=(20,17))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd

# Assuming 'data_cleaned' is your DataFrame and 'numerical_df' contains numerical columns

# Calculate correlation matrix for the numerical columns
corr_matrix = numerical_df.corr()

# Set the correlation threshold
threshold = 0.8

# Find pairs of columns with high correlations (greater than threshold)
high_corr_var = np.where(corr_matrix > threshold)

# List to store pairs of columns with high correlations
high_corr_pairs = [(numerical_df.columns[x], numerical_df.columns[y])
                   for x, y in zip(*high_corr_var)
                   if x != y and x < y]

# Create a set of columns to remove (using col2 to avoid duplicates)
columns_to_remove = set()
for col1, col2 in high_corr_pairs:
    columns_to_remove.add(col2)  # Modify if you want to remove col1 instead

# Convert set to list for further use
columns_to_remove = list(columns_to_remove)

# Drop the redundant columns from the cleaned data
df_cleaned_corr = new_Train.drop(columns=columns_to_remove)

# Print the columns removed and preview the cleaned DataFrame
print("Columns removed:", columns_to_remove)
print(df_cleaned_corr.head())
```
